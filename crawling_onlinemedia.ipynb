```python
#crawling and scraping online media
#TOI 

"Storing content and image urls of midas web pages in .txt files and .json files"""
import json

import requests
from bs4 import BeautifulSoup

from urllib.request import urlopen 
from bs4 import BeautifulSoup

class Web_urls(object):
    def __init__(self,url):
        self.wlink = requests.get(url)
        print(self.wlink.content)

    def content_bs(self):
        print("\n\n\n\nafter using Beautifulsoup library,the parsed html content is:\n\n\n")
        self.soup = BeautifulSoup(self.wlink.content,'html5lib')
        print(self.soup.prettify)
        
        
    def url_images(self):   
        print("\n\n\n\nThe list of images(with <img>) in web page are:\n\n\n")
        images=self.soup.find_all('img')
        #print(images)
        
        
             
    
           
```

```python
c=Web_urls("https://timesofindia.indiatimes.com/life-style/health-fitness/de-stress/google-honours-dr-herbert-kleber-here-is-what-you-need-to-know/articleshow/71387954.cms")                       
c.content_bs()
c.url_images()
```

```python
for im in c.url_images().images:
    print(im["src"]+"\n")
```

```python
fpgtext=c.soup.findAll("p")
fpgtext_a=c.soup.findAll("a")
fpgtext_art=c.soup.find("arttitle")
fpgtext_div= c.soup.find("div",attrs={'class':'Normal'})
fpgtext_date= c.soup.find("div",attrs={'data-plugin':'dateformat'})
fpgtext_smedia = c.soup.findAll("i")
fpgtext_ads= c.soup.find("img", attrs ={'alt':'Advertisement'})
images=c.soup.findAll("img")
source_date_time=c.soup.find('div',attrs={"id":"adsdivLyr"})
content =fpgtext_div.get_text()

print("\n\n it's p tag:",fpgtext ,"\n\n\n\n\n")
print("\n\n\n\n\n it's a tag:",fpgtext_a,"\n\n\n\n\n")
print("\n\n\n\n\nits art tag:",fpgtext_art.get_text() ,"\n\n\n\n\n")
print("\n\ndiv tag:its div tag:\n\n",fpgtext_div)
print("\n\ndate:",fpgtext_date.get_text())
#print("\n\nsocial medai:",fpgtext_smedia.get_text())
print("\n\nits img tag ads :", fpgtext_ads)
print("\n\nhref:", c.soup.div.a.href)
print(source_date_time.get_text())

print("\n\n\n\n\n list of a tag:\n\n",fpgtext_a)
for txt in fpgtext_a :
    print(txt.find(attrs= {"data-section":"46551404"}),"\n")

print("\n\nP tag:\n")
for txt in fpgtext :
    print(txt.get_text())

print("\n\nlist with text of a tag:")
for txt_a in fpgtext_a:
    print(txt_a.get_text())

print("\n\narticle content:\n",content)
    
print("\n\nad images:")
for i in fpgtext_a :
    print( i,"\n")
    
# regular expression
        
```

```python
category = c.soup.find_all("a" ,attrs= {"data-action":"Navigation_MenuItem"})
for cat in category:
    print(cat.get_text() ,"\n")
```

```python
my_cat= c.soup.find_all("span",attrs={"itemprop":"name"})
for ct in my_cat:
    print(ct.get_text())
```

```python
nav = c.soup.find_all("a" ,attrs= {"data-action":"Navigation_NewsItem"})
for n in nav:
    print(n.get_text() ,"\n")
```

```python
source = c.soup.find("a",attrs = {"name":"etimeshomelink"})
url_source=source["href"]
print(url_source)
toi= c.soup.find("div",attrs={"class":"toilink"})
toi_url = toi.find("a")["href"]
print(toi_url)
```

```python
fpgtext_toi = c.soup.find("div",attrs={"class":"as_byline"})
sm =fpgtext_toi.find_all("i")
for s in sm:
    print(s.get_text())


other= c.soup.find_all("img",attrs={"data-plugin":"imagesection1"})
for o in other:
    print(o["src"])
    
for i in range(1,10):
    print(c.soup.find("a", attrs={"data-pos":"{}".format(i)}).get_text())
```

```python
with open("toi_article.txt","w",encoding = "UTF-8") as f:
    f.write("\n\tSource:\n"+source.get_text())
    f.write("\n\n\tTOI url: "+toi_url+"\n\n\tURL of source page: "+url_source)
    
    f.write("\n\n\tList of  all categories:\n")
    for cat in category:
        f.write(cat.get_text()+"\n")
    
    f.write("\n\n\tCurrently in category:\n")
    for ct in my_cat:
        f.write(" >> "+ct.get_text())
    f.write("\n\n\tdate:\n"+ fpgtext_date.get_text())
    f.write("\n\n\n\n\n\tArticle heading:  " + fpgtext_art.get_text() +"\n\n\n\n\n")
    f.write("\tArticle content:\n\t"+content+"\n\n")
    f.write("\tOther news list:\n")
    for i in range(1,10):
        f.write(c.soup.find("a", attrs={"data-pos":"{}".format(i)}).get_text()+"\n")
    
    f.write("\n\n\tOther news from all categories:\n")
    for index,n in enumerate(nav):
        f.write("{}). ".format(index) +n.get_text() +"\n")

    f.write("\n\tList of image urls:\n")
    for index,im in enumerate(images) :
        f.write("{}). ".format(index) + im["src"]+"\n")
    f.write("\n\n\tTOI is on following social media:\n")
    for s in sm:
        f.write("--"+s.get_text()+"\n")
    
```
